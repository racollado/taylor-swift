```python
from bs4 import BeautifulSoup
import requests
import csv

genres = "pop-songs", "country-airplay-songs", "dance-mix-show-airplay-songs", "hot-100-songs", "christian-airplay-songs"
years = "2020", "2019", "2017", "2014", "2010", "2008", "2006"
all_titles = ['Song Title']
all_artists = ['Artist']
all_genres = ['Genre']
all_years = ['Year']


for genre in genres:
    
    for year in years:

        response = requests.get("https://www.billboard.com/charts/year-end/"+ year + "/" + genre + "/")
        html_string = response.text
        soup = BeautifulSoup(html_string, 'html.parser')

        # let's isolate just the song titles
        html_we_want = soup.find_all("li", class_="o-chart-results-list__item // lrv-u-flex-grow-1 lrv-u-flex lrv-u-flex-direction-column lrv-u-justify-content-center lrv-u-border-b-1 lrv-u-border-color-grey-light lrv-u-padding-l-2 lrv-u-padding-l-1@mobile-max")
        titles = soup.find_all("h3", id="title-of-a-story")      
        artists = soup.find_all("span", class_=["c-label a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block", "c-label a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block u-font-size-20@tablet"])
        
    #still missing the first artist!!
        i = 0
        while i < 10:
#replace spaces with dashes in titles and artists      
#cleaning up titles!!
#replace '-/-' , '-x-', '-&-', '-+-' with '-and-'
#replace '$' with 's'
#remove ',' and ''' <--!!last one not done! (rare enough to do manually)
#trim '-Featuring....' and '-(...'
            
            artist = artists[i].string.strip().replace(" ", "-").replace(".", "").replace('-/-', '-and-').replace('-x-', '-and-').replace('-&-', '-and-').replace('-+-', '-and-').replace('-with-', '-and-').replace('$','s').replace(',', '').replace('!', '')
            if '-Feat' in artist:
                artist = artist.split('-Feat')[0]
            if '-Duet' in artist:
                artist = artist.split('-Duet')[0]
            all_artists.append(artist)
            
            title = titles[i].string.strip().replace(" ", "-").replace(',', '')
            if '-(' in title:
                title = title.split('-(')[0]
            all_titles.append(title)
            
            genretag = genre.split('-')
            all_genres.append(genretag[0])
            
            all_years.append(year)
          
            i += 1    
    
with open("BillboardYearEndTop10.csv", 'w') as file:
    writer = csv.writer(file)
    writer.writerow(all_titles) 
    writer.writerow(all_artists) 
    writer.writerow(all_genres) 
    writer.writerow(all_years) 

#manually remove repeats for 2006, etc where there is no data

```
